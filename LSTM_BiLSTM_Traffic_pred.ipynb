```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
# Define the length of prediction period (1 day = 288 intervals for 5-minute data)
prediction_period = 30 * 288

# Split the data
train_data = data[:-prediction_period]
test_data = data[-prediction_period:]
```

```python
import numpy as np

def create_data(sq):
    def create_sequences(data, seq_length):
        xs, ys = [], []
        for i in range(len(data) - seq_length):
            x = data[i:i+seq_length]
            y = data[i+seq_length]
            xs.append(x)
            ys.append(y)
        return np.array(xs), np.array(ys)

    # Define sequence length (e.g., 2 hours = 24 intervals for 5-minute data)
    seq_length = sq

    # Create sequences
    X_train, y_train = create_sequences(train_data, seq_length)
    X_test, y_test = create_sequences(test_data, seq_length)
    return X_train, X_test, y_train, y_test
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

def build_model(sq, op):
    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(128, return_sequences=True, input_shape=(sq, 325)))
    model.add(Dropout(0.3))
    model.add(LSTM(64, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(32))
    model.add(Dense(325))
    # Compile the model
    model.compile(optimizer=op, loss='mean_squared_error')
    return model
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional

def build_model_a(sq, op):
    # Build the LSTM model
    model = Sequential()
    model.add(Bidirectional(LSTM(128, return_sequences=True, input_shape=(sq, 325))))
    model.add(Dropout(0.3))
    model.add(Bidirectional(LSTM(64, return_sequences=True)))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(64, return_sequences=True)))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(64, return_sequences=True)))
    model.add(Dropout(0.2))
    model.add(Bidirectional(LSTM(64, return_sequences=True)))
    model.add(Dropout(0.2))
    model.add(LSTM(32))
    model.add(Dense(325))
    # Compile the model
    model.compile(optimizer=op, loss='mean_squared_error')
    return model
```

```python
def model_performance(history):
    # Extract loss and metrics
    history_dict = history.history
    epochs = range(1, len(history_dict['loss']) + 1)

    # Plot training and validation loss
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, history_dict['loss'], 'b', label='Training loss')
    plt.plot(epochs, history_dict['val_loss'], 'r', label='Validation loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    return plt
    
```

```python
def make_predictions(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)

    # Inverse transform the predictions and true values
    predictions = scaler.inverse_transform(predictions)
    y_test = scaler.inverse_transform(y_test)
    return predictions, y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluation(y_test, predictions):
    mse = mean_squared_error(y_test, predictions)

    # Calculate RMSE
    rmse = np.sqrt(mse)

    # Calculate MAE
    mae = mean_absolute_error(y_test, predictions)

    # Calculate R-squared
    r2 = r2_score(y_test, predictions)
    return mse, rmse, mae, r2
```

```python
# Plot the predictions - for the sensor
import matplotlib.pyplot as plt

def plot_prediction(y_test, predictions):
    plt.figure(figsize=(10, 6))
    plt.plot(y_test[: , 0], label='True Value')
    plt.plot(predictions[: , 0], label='Predicted Value')
    plt.legend()
    return plt
```

```python
def plot_timestamp(y_test, predictions):
    # For a timestamp
    plt.figure(figsize=(10, 6))
    plt.plot(y_test[0, :], label='True Value')
    plt.plot(predictions[0, :], label='Predicted Value')
    plt.legend()
    return plt
```

#### K-Fold

```python
from sklearn.model_selection import KFold
```

```python
# Define the number of folds
k = 5

# Initialize KFold
kf = KFold(n_splits=k, shuffle=True, random_state=42)
```

```python
# Initialize lists to store results
mse_list, rmse_list, mae_list, r2_list = [], [], [], []
```

```python
def build_model_k():
    # Build the LSTM model
    model_k = Sequential()
    model_k.add(LSTM(128, return_sequences=True, input_shape=(sq, 325)))
    model_k.add(Dropout(0.3))
    model_k.add(LSTM(128, return_sequences=True))
    model_k.add(Dropout(0.2))
    model_k.add(LSTM(64, return_sequences=True))
    model_k.add(Dropout(0.2))
    model_k.add(LSTM(32))
    model_k.add(Dense(325))
    # Compile the model
    model_k.compile(optimizer='adam', loss='mean_squared_error')
    return model_k
```

```python
sq = 24

X_train, X_test, y_train, y_test = create_data(sq)
```

```python
initial_model = build_model_k()
initial_weights = initial_model.get_weights()
```

```python
epo = 1
for train_index, val_index in kf.split(X_train):
    print('\n----KFold '+str(epo)+'---\n')
    epo += 1
    # Clear the session to free memory
    K.clear_session()

    # Rebuild the model
    model = build_model_k()

    # Reset the model weights to initial weights
    model.set_weights(initial_weights)


    # Split the data into training and validation sets
    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Train the model
    model.fit(X_train_fold, y_train_fold, epochs=20, batch_size=128, validation_data=(X_val_fold, y_val_fold), verbose=1)

    # Save the trained model weights
    model.save_weights('temp_model_weights_k.h5')

    # Clear the session to free memory after training
    K.clear_session()
    gc.collect()

    # Rebuild the model and load the trained weights for prediction
    model = build_model_k()
    model.load_weights('temp_model_weights_k.h5')


    # Make predictions
    val_predictions = model.predict(X_val_fold)

    # Inverse transform the predictions and true values
    val_predictions = scaler.inverse_transform(val_predictions)
    y_val_fold_p = scaler.inverse_transform(y_val_fold)

    # Calculate evaluation metrics
    mse = mean_squared_error(y_val_fold_p, val_predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_val_fold_p, val_predictions)
    r2 = r2_score(y_val_fold_p, val_predictions)
    print(r2)

    # Store the results
    mse_list.append(mse)
    rmse_list.append(rmse)
    mae_list.append(mae)
    r2_list.append(r2)
```

```python
r2_list
```

```python
# Calculate average metrics
avg_mse = np.mean(mse_list)
avg_rmse = np.mean(rmse_list)
avg_mae = np.mean(mae_list)
avg_r2 = np.mean(r2_list)

print(f'Average MSE: {avg_mse}')
print(f'Average RMSE: {avg_rmse}')
print(f'Average MAE: {avg_mae}')
print(f'Average R^2 Score: {avg_r2}')
```
